We see a high amount of skew in the firm data, which suggests we may need to transform the data in order to have more normally distributed data (which helps with meaningful analysis). Let's try out two different transformations and see how they compare.

```python
# first, let's try a Box-Cox transformation
# the Box-Cox transform is used in advanced data anslysis and attempts to normalize the data 
boxed_firm_dat = pd.DataFrame(firm_dat, copy=True, columns = firm_dat.keys())
lambda_list = pd.Series(index=firm_dat.keys())

# Sparsity issues necessitate a two-parameter Box-Cox transformation
# more specifically, we add an alpha term of 1 so that zero entries can still be transformed
# this does destroy sparsity, but we can re-introduce it later
boxed_firm_dat += 1
for i in boxed_firm_dat.columns:
    boxed_firm_dat[i], lambda_list.loc[i] = boxcox(boxed_firm_dat[i])

# we add a one row, two column format to see the results side-by-side
fig, (ax, ax2) = plt.subplots(ncols=2, sharey=True, figsize=(18,6))
plt.tight_layout()

# add the first plot, of the Box-Cox transform 
__ = boxed_firm_dat.plot.box(ax = ax)

# Our second transformation divides the data by the maximum value, which makes all of the data between zero and one
# this technique preserves sparsity and can ensure all of the features are on the same scale
# to make this more normal, we also take a square root, which reduces the variance and skew
maxAbs = MaxAbsScaler(copy=True)
sqAbs_firm_dat = pd.DataFrame(np.sqrt(maxAbs.fit_transform(firm_dat)), columns=firm_dat.keys())

# add the second plot,of the SqAbsScaled data
__ = sqAbs_firm_dat.plot.box(ax = ax2)
```

The two transformations fail to have much normality as the data is incredibly sparse. At least we have managed to create a few different transformations of our data that may help us in clustering. We may also want to examine the collinearity of the data to examine the dimensionality of the data before we begin to cluster. To do this, we'll use a common indicator of collinearity, the variance inflation factor, or VIF.

We also need to add a column of ones to serve as the intercept in Python, so the last VIF for each dataframe won't be very meaningful.

```python
print('VIFs for the cleaned, original data: ')
vifs_firm_dat = [vif(firm_dat.assign(ones=1).as_matrix(),i) for i in range(firm_dat.shape[1])]
print(np.round(vifs_firm_dat, 2))

print('VIFs for the Box-Cox data: ')
vifs_boxed_dat = [vif(boxed_firm_dat.assign(ones=1).as_matrix(),i) for i in range(boxed_firm_dat.shape[1])]
print(np.round(vifs_boxed_dat,2))

print('VIFs for the SqAbsScaled data: ')
vifs_sqAbs_dat = [vif(sqAbs_firm_dat.assign(ones=1).as_matrix(),i) for i in range(sqAbs_firm_dat.shape[1])]
print(np.round(vifs_sqAbs_dat,2))
```

*Sacrebleu!*

That's a lot of collinearity in the original data. A variance inflation factor, or VIF, is a measure of the similarity between features in a dattaset. Basically, it tells us if any columns can be recreated by other columns and are thus redundant (or creating singular matrices). Anything above 5 is regarded as collinear, and VIFs much higher than that likely necessitate processing to ensure a the data is usable.

The untransformed data has a smallest VIF os 9.49! (the last VIF is the intercept column) This means that we likely are looking at nearly single-dimensional data and even after a collinearity reduction technique such as Principal Components Analysis, or PCA, which can also reduce the dimensionality to improve clusterability.

The Box-Cox is very promising, although we still have two columns with VIFs just over 5, so we may examine clustering with and without PCA.

The SqAbsScaled data is much better than the original data, but we still notice a high number of VIFs over 5, so we should definitely use PCA.

```python
np.set_printoptions(precision = 4, suppress=True)

pca_firm = PCA(n_components=2)
pca_firm.fit(firm_dat)
pca_firm.explained_variance_ratio_
```

```python
pca_firm_boxed = PCA(n_components=2)
pca_firm_boxed.fit(boxed_firm_dat)
pca_firm_boxed.explained_variance_ratio_
```

```python
pca_firm_sqAbs = PCA(n_components=2)
pca_firm_sqAbs.fit(sqAbs_firm_dat)
pca_firm_sqAbs.explained_variance_ratio_
```

Principle component analysis, or PCA, is statistical technique that simplifies data (so it's easier to visualize and to compute) and also helps us solve structural issues like collinearity. What we see above is the estimate of how well we can express the data in reduced dimensions, namely 2 dimensions for each of the datasets above.

Our original dataset is certainly collinear: a single dimension explains over 99.9% of the variation in the data - basically it's all just one straight line.

By transforming our data, we do lose some of the predictability, but this likely also increases clusterability and predictability later. Now, let's examine the components for each dataset and see how transformations affected the components.

**Technical Note:** PCA is not robustified to sparsity, and while recent developments in Robust PCA methods exist (such as Candes, Li, Ma, Wright (2009)), we consider them beyond the scope of this project as they require sophisticated knowledge of convex optimization and are not implemented in Sci-Kit Learn or other typical Python modules as of yet. Thus, we should not consider this PCA to be a prefect projection of our data, although it allows us to cluster and predict to a sufficient degree of accuracy to be interesting.  

```python
pca_results = vs.pca_results(firm_dat, pca_firm)
print(pca_results.cumsum())
```

Here we see interesting weightings on each PCA. 

The original data depended largely on the first two features, 'Unknown Size' and 'Size 1-5'. The Box-Cox transformed data depends less on these two, but weights each larger business segment slightly less than each smaller segment. Finally, we see that in the SqAbsScaled data, most of the weight is on the medium and even larger sized companies, which has b een scaled to the same size as the smaller companies. 

This reveals that PCA is sensitive to scale, and that we can create substantially different component analyses of the dame data by transforming it in simple steps.

It should be noted that the transformed data predict with quite a bit less explained variance - this is likely because we have proportionally increased the variance in the larger company categories by scaling the data, whereas before, the smaller companies contained most of the variation as they had a much higher range.

**Heads up:** This code may take 2-3 minutes on a < 4 core machine

```python
# The next couple of steps create dicts for us to recycle code

# Here we create a list of columns and clustering algos
data_cols = ['Orig', 'PCA Orig','Box-Cox','PCA BxCx','PCA SqAbs']
cluster_algos = [GaussianMixture, KMeans]

# Here we create our PCA'd data and store all of the data in a dict for ease of use
pca_firm_dat = pca_firm.transform(firm_dat)
pca_boxed_firm_dat = pca_firm_boxed.transform(boxed_firm_dat)
pca_sqAbs_firm_dat = pca_firm_sqAbs.transform(sqAbs_firm_dat)

datasets = dict(zip(data_cols,[firm_dat,pca_firm_dat,boxed_firm_dat,pca_boxed_firm_dat,pca_sqAbs_firm_dat]))

# we have some dimensinoality issues with the PCA Orig data, so we only cluster for 3+ clusters
t = range(2,10)
cluster_range = dict(zip(data_cols,[t,range(3,10),t,t,t]))

"""    # Here we create two dataframes to store 
    data_index = [i for i in range(2,10)]
    gmm_data_scores = pd.DataFrame(np.zeros((8,5)),index=sil_index,columns=sil_cols) 
    kmeans_data_scores = pd.DataFrame(np.zeros((8,5)),index=sil_index,columns=sil_cols) 
"""

# these hold our predictions and centers
gmm_preds = {}
gmm_centers = {}
kmeans_preds = {}
kmeans_centers = {}

# place our preds and centers into a dict 
preds = dict(zip(cluster_algos,[gmm_preds,kmeans_preds]))
centers = dict(zip(cluster_algos,[gmm_centers,kmeans_centers]))

# GaussianMixture and KMeans have different arguments and attributes, so we need dicts for that
count = dict(zip(cluster_algos,['n_components','n_clusters']))
attributes = dict(zip(cluster_algos,['means_','cluster_centers_']))

for dat in data_cols:
    for algo in cluster_algos:
        for i in cluster_range[dat]:
            
            # let's slice those dicts into easier to read formats 
            data = datasets[dat]
            preds_dict = preds[algo]
            cluster_dict = centers[algo]
            cluster_count = count[algo]
            
            # here we deal with the different arguments and attributes
            arguments = {cluster_count:i,'random_state':1920}
            centroid_attribute = attributes[algo]
            
            # here we fit the clusterer with each algo from cluster_algos
            clusterer = algo(**arguments)
            clusterer.fit(data)
            
            # here we add the preds and centers to our dicts of preds and center for each algo
            preds_dict[(dat,i)] = clusterer.predict(data) 
            cluster_dict[(dat,i)] = getattr(clusterer, centroid_attribute)
```

```python
# let's make some frames to hold and format our output
sil_cols = ['PCA Orig','Box-Cox','PCA BxCx','PCA SqAbs']
sil_index = [i for i in range(2,10)]

sil_scores = pd.DataFrame(np.zeros((8,4)),index=sil_index,columns=sil_cols) 

# let's transform and cluster the original data 
# we only really need the first dimension since it's 99.9% of the variation
pca_firm_dat = pca_firm.transform(firm_dat)
for i in range(3,10):
    clusterer = GaussianMixture(n_components = i, random_state=1920)
    clusterer.fit(pca_firm_dat)
    preds = clusterer.predict(pca_firm_dat)
    centers = clusterer.means_

    score = silhouette_score(pca_firm_dat, preds, sample_size= 10000, random_state=1920)

    sil_scores['PCA Orig'].loc[i] = score

# let's try clustering the Box-Cox data without PCA
# this data has more collinearity, but we also should have more information about each point
for i in range(2,10):
    clusterer = GaussianMixture(n_components = i, random_state=1920)
    clusterer.fit(boxed_firm_dat)
    preds = clusterer.predict(boxed_firm_dat)
    centers = clusterer.means_

    score = silhouette_score(boxed_firm_dat, preds, sample_size= 10000, random_state=1920)

    sil_scores['Box-Cox'].loc[i] = score

pca_boxed_firm_dat = pca_firm_boxed.transform(boxed_firm_dat)
for i in range(2,10):
    clusterer = GaussianMixture(n_components = i, random_state=1920)
    clusterer.fit(pca_boxed_firm_dat)
    preds = clusterer.predict(pca_boxed_firm_dat)
    centers = clusterer.means_

    score = silhouette_score(pca_boxed_firm_dat, preds, sample_size= 10000, random_state=1920)

    sil_scores['PCA BxCx'].loc[i] = score
    
pca_sqAbs_firm_dat = pca_firm_sqAbs.transform(sqAbs_firm_dat)
for i in xrange(2,10):
    clusterer = GaussianMixture(n_components = i, random_state = 1920)
    clusterer.fit(pca_sqAbs_firm_dat)
    preds = clusterer.predict(pca_sqAbs_firm_dat)
    centers = clusterer.means_

    score = silhouette_score(pca_firm_dat, preds, sample_size= 10000, random_state = 1920)

    sil_scores['PCA SqAbs'].loc[i] = score

print(sil_scores)
```

Above we calculated the 'silhouette score,' a useful metric for examining how 'clusterable' the data is for some number of clusters (i represents the number of clusters in the above code). A higher silhouette score suggests a good fit for the number of clusters we have selected, and the score varies from +1 to -1. Due to the large sample size, we subsample 10,000 points to ease the computational speed of the algorithms. 

Typically, the clusterability decreases as size increases, but sometimes we see local maxima as the size increases. We have a zero entry for the 'PCA Orig' x 2 point as the linearity of the original data frustrates the clusterability (it ends up assigning all of the data to a single cluster, which flips an error in the silhouette scorer). Each of the columns is substantially different than the others, which is due in part to dimensionality as well as the structural properties of each of the datasets. 

For the 'PCA Orig,' we could justify a number of clusters, ideally 3 clusters, but 4 and 5 clusters still have a relatively high silhouette score. The 'Box-Cox' data without PCA can only really hold under 2 clusters, but still clusters poorly, although the 'PCA BxCx' is a good candidate for two clusters. 'SqAbs' can easily be grouped in either 2 or 3 clusters while still maintaining decent clusterability. Let's start with these enumerated examples and a few with more clusters and see how they visualize and why the lower cluster counts cluster better.

We see from the 'Box-Cox' $\rightarrow $ 'PCA BxCx' that PCA dramatically improved clusterability for all of the listed number of clusters. This is largely due to dimensionality reduction.

```python
# create a dict of cluster counts for each of our datasets
clusts = {}

cluster_list = ['PCA Orig', 'PCA BxCx', 'PCA SqAbs']

# iterate over our dataset categories in the sil_cols var
clusts[cluster_list[0]] = [3, 4, 5]
clusts[cluster_list[1]] = [2, 3, 4]
clusts[cluster_list[2]] = [2, 3, 4]

# create a dict for our dataset names
dat_list = {}
dat_list[cluster_list[0]] = pca_firm_dat
dat_list[cluster_list[1]] = pca_boxed_firm_dat
dat_list[cluster_list[2]] = pca_sqAbs_firm_dat

# create a prediction tracking dictionary for later
pred_list = {}

# create a center-storing database for later
center_list = {}

# create a nice figure
fig, ((ax1, ax2, ax3), (ax4, ax5, ax6), (ax7, ax8, ax9)) = plt.subplots(nrows=3, ncols=3, sharey= 'row', figsize = (18,24))
ax_list = [ax1, ax2, ax3, ax4, ax5, ax6, ax7, ax8, ax9]
ax_count = 0
cmap = cm.get_cmap('gist_rainbow')


for i in cluster_list:    
    data = pd.DataFrame(dat_list[i], columns=['Dimension 1', 'Dimension 2'])
    
    for j in clusts[i]:
        clusterer = GaussianMixture(n_components = j, random_state = 1920)
        clusterer.fit(data)
        pred_list[(i,j)] = clusterer.predict(data)
        center_list[(i,j)] = clusterer.means_[:,0:2]

        predictions = pd.DataFrame(pred_list[(i,j)], columns = ['Cluster'])
        plot_data = pd.concat([predictions, data.iloc[:,0:2]], axis = 1)

# Color the points based on assigned cluster
        for k, cluster in plot_data.groupby('Cluster'):   
            cluster.plot(ax = ax_list[ax_count], kind = 'scatter', x = 'Dimension 1', y = 'Dimension 2', \
                         color = cmap((k)*1.0/(len(center_list[(i,j)])-1)), label = 'Cluster %i'%(k), s=30);
    
        for k, c in enumerate(center_list[(i,j)]):
            ax_list[ax_count].scatter(x = c[0], y = c[1], color = 'white', edgecolors = 'black', \
                       alpha = 1, linewidth = 2, marker = 'o', s=200);
            ax_list[ax_count].scatter(x = c[0], y = c[1], marker='$%d$'%(k), alpha = 1, s=100);
            
        ax_count += 1
```

In the plots above, we plot the following data in descending order: 'PCA Orig', 'PCA BxCx', and 'PCA SqAbs'. Horizontally, we increase the number of clusters and we can observe the color changes that correspond to new labels in each cluster.

If you're saying *sent le poisson*, yes something does seem suspicious. The 'PCA Orig' looks nearly identical in its clustering - is anything even assigned to some of the clusters? Let's examine the cluster counts below:

```python
cluster_analysis_index = list()

for i in cluster_list:
    for j in clusts[i]:
        cluster_analysis_index.append((i,j))

pred_unique = {}
pred_count = {}

print '\033[1m' + '(Data, # Cluster)\t' + 'Clusters: Count\t' + '\033[0m'
for i in cluster_analysis_index:
        out = ''
        out += str(i) + ' \t'
        pred_unique[i] = np.unique(pred_list[i])
        for j in pred_unique[i]:
            pred_count[(i,j)] = sum(pred_list[i]==j)
            out += str(j) + ': ' + str(pred_count[(i,j)]) + '    \t'
        print out
            
```

Here we see that many of the 'good' clusterings aren't all that spectacular - the 'PCA Orig' returns a cluster of size < 6% in all of the clusters above, which suggests that the data really may not belong in multiple clusters at all (we may just have some outlier or a bad clustering). The 'PCA Orig', 5 even has a single-point cluster! This is likely some outlier we missed earlier. This shows us that sihouette score isn't the ultimate metric, and we should always examine our clusters more closely. 

Hence, if we were to choose a clustering to rely on, we should look to the 'PCA BxCx' dataset, as it visually makes sense for the data and still has an OK silhouette score. 2 clusters is probably ideal, but we could go with 3 or maybe even 4, given the smooth parsing of the data they seem to provide. Let's try examining the cluster characteristics for all three below, as well as for the first and second 'SqAbs' since they may turn out worthwhile (the second has one cluster with only 7.3% of the points in it, but that's probably large enough for our purposes).

**Technical note:** The Box-Cox transform, while it failed to truly normalize the data in the larger features (the 'Size 500+' still looks like a Bernoulli distribution), did help us find the 'most normal' transformation of our data and synergizes well with Gaussian Mixture Models, which assume (independent) normally distributed features. Note that both of our transformations do destroy much of the covariance between features as they scale each feature independently. In addition to selecting different transformations, should we have chosen a different clustering algorithm or utilized a different distance metric such as Manhattan distance, we would have drastically different clustering results and likely see at least one far superior clustering for a number of metrics. For sake of simplicity, we stick to these two relatively easy to visualize, justify, and code transformations, but the nature of sparse data analysis merits far more sophisticated statistical analysis. 

```python
# let's create an inverse function for the Box-Cox to examine our centers
def inverse_boxcox(y, lam):
    if(lam == 0):
        return np.exp(y.values) - 1
    else:
        return np.power(y.values*lam + 1, 1/lam) - 1

# store our means for each (dataset, #cluster) combo
cluster_means = {}   
    
for i in cluster_analysis_index[3:6]:
    # which (dataset, #cluster) combo?
    print(i)
    
    # grab our centers from our list
    centers = center_list[i]
    
    # create a DataFrame for each of our centers
    bxcx_centers = pca_firm_boxed.inverse_transform(centers)
    bxcx = pd.DataFrame(bxcx_centers, columns=boxed_firm_dat.keys())
    
    # invert the centers from the Box-Cox transform
    for j in bxcx.columns:
        bxcx[j] = np.round(inverse_boxcox(bxcx[j],lambda_list[j]),2)

    # create a column to see the counts for each cluster
    bxcx['Count'] = np.zeros(bxcx.shape[0])
    for j in bxcx.index:
        bxcx['Count'].loc[j] = pred_count[(i,j)]
    
    # save and output our cluster means
    cluster_means[i] = pd.DataFrame(bxcx, copy=True)
    display(bxcx)

# now let's do the SqAbs data
for i in cluster_analysis_index[6:8]:
    print(i)
    
    centers = center_list[i]
    
    # create a DataFrame for each of our centers
    sqAbs_centers = pca_firm_sqAbs.inverse_transform(centers)
    sqAbs = sqAbs_centers
    # invert the centers from the MaxAbs transform
    sqAbs = pd.DataFrame(np.round(maxAbs.inverse_transform(np.square(sqAbs)),2),columns=sqAbs_firm_dat.keys())

    # create a column to see the counts for each cluster
    sqAbs['Count'] = np.zeros(sqAbs.shape[0])
    for j in sqAbs.index:
        sqAbs['Count'].loc[j] = pred_count[(i,j)]
    
    # save and output our cluster means
    cluster_means[i] = pd.DataFrame(sqAbs, copy=True)
    display(sqAbs)
    
```

From the output above, we can't actually tell too much about our clusters, other than they seem to follow a high business count/small business count division in every clustering.

```python
common_full_pop = pop_dat.loc[common_index.values]

for i in cluster_analysis_index:
    labels = pred_list[i]

    # create a set of filtered labels by adding them to the firm data, then slicing only common labels
    filtered_labels = pd.DataFrame(firm_dat, copy=True)
    filtered_labels['Labels'] = labels
    filtered_labels = filtered_labels.loc[common_index]

    common_full_pop['Labels'] = pd.Series(filtered_labels['Labels'], copy=True)

    classer = ABC(DTC(max_depth = 3), n_estimators = 60)

    full_X_train, full_X_test, full_y_train, full_y_test = TTS(common_full_pop.iloc[:,0:-1], common_full_pop.loc[:,'Labels'], test_size = 0.1)
    
    classer.fit(full_X_train, full_y_train)
    
    score = np.round(classer.score(full_X_test, full_y_test),3)
    
    print('Clustering: '+  str(i) + ' boosted decision tree accuracy: ' + str(score) + '\n Most important features: (importance)')
    
    features = pd.Series(classer.feature_importances_, index = common_full_pop.columns[0:-1])
    
    features.sort_values(ascending = False, inplace=True)
    
    display(features.head(5))
```
